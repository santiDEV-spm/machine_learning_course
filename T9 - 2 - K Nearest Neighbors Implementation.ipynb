# CREANDO NUESTRO KNN

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
import warnings
from math import sqrt
from collections import Counter
```

```python
dataset = {
    'k':[[1,2],[2,3],[3,1]],
    'r':[[6,5],[7,7],[8,6]]
}
new_point = [5,7]
```

```python
[[plt.scatter(ii[0],ii[1],s=50, color = i) for ii in dataset[i]]for i in dataset]
plt.scatter(new_point[0],new_point[1],s=100)
```

```python
def K_nearest_neighbors(data,predict, k = 3, verbose=False):
    
    if len(data)>=k:
        warnings.warn('K es un valor menor al numero de elementos a votar')
    
    distances = []
    for group in data:
        for feature in data[group]:
           # d = sqrt((feature[0]-predict[0])**2 + (feature[1]-predict[1])**2)
           # d = np.sqrt(np.sum((np.array(feature) - np.array(predict))**2))
            d = np.linalg.norm(np.array(feature) - np.array(predict))
            distances.append([d,group])
    if verbose:
        print(distances)        
        print(voutes)
        print(vote_result)
    voutes = [i[1] for i in sorted(distances)[:k]]
   
    
    vote_result = Counter(voutes).most_common(1)
   
    
    return vote_result[0][0]
```

```python
result = K_nearest_neighbors(dataset, new_point)
result
```

```python
[[plt.scatter(ii[0],ii[1],s=50, color = i) for ii in dataset[i]]for i in dataset]
plt.scatter(new_point[0],new_point[1],s=100, color = result)
```

# Aplicando nuestro KNN al dataset del Cancer 

```python
import pandas as pd
```

```python
df = pd.read_csv('../python-ml-course-master/python-ml-course-master/datasets/cancer/breast-cancer-wisconsin.data.txt')
```

```python
df.replace('?', -99999, inplace=True)
```

```python
df.columns = ['name', 'V1','V2','V3', 'V4', 'V5','V6', 'V7','V8','V9', 'CLASS']
df.drop(['name'],1,inplace=True)
```

```python
df.head()
```

```python
full_data = df.astype(float).values.tolist()
```

```python
full_data
```

```python
import random
```

```python
random.shuffle(full_data)
```

```python
test_size = 0.2
```

```python
train_set = {2:[],4:[]}
test_set = {2:[],4:[]}
```

```python
train_data = full_data[:-int(test_size*len(full_data))]
test_data = full_data[-int(test_size*len(full_data)):]
```

```python
for i  in train_data:
    train_set[i[-1]].append(i[:-1])
for i in test_data:
    test_set[i[-1]].append(i[:-1])
```

```python
train_set
```

```python
correct = 0
total = 0
for group in test_set:
    for data in test_set[group]:
        vote = K_nearest_neighbors(train_set,data, k=5)
        if group == vote:
            correct +=1
        total += 1
print('Correctos', correct)
print('Total', total)
print('eficiencia del knn = ', correct/total)
```
