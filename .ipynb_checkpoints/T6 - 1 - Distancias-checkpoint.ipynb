# Distancias

```python
from scipy.spatial import distance_matrix
import pandas as pd
```

```python
data = pd.read_csv("../python-ml-course-master/python-ml-course-master/datasets/movies/movies.csv", sep=';')
data
```

```python
movies = data.columns.values.tolist()[1:]
movies
```

```python
dd1 = distance_matrix(data[movies], data[movies], p=1)#DISTANCIA MANHATTAN
dd2 = distance_matrix(data[movies], data[movies], p=2)#EUCLIDEA
dd10 = distance_matrix(data[movies], data[movies], p=10)#MINKOSKI
```

```python
def dm_to_df(dd, col_name):
    import pandas as pd
    return pd.DataFrame(dd, index=col_name, columns=col_name)
```

```python
dm_to_df(dd1, data['user_id'])
```

```python
dm_to_df(dd2,data['user_id'])
```

```python
dm_to_df(dd10,data['user_id'])
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig= plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.scatter(xs=data['star_wars'],ys=data['lord_of_the_rings'],zs=data['harry_potter'])
```

# Clustering Jerarquico

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
movies
```

```python
data[movies]
```

```python
Z = linkage(data[movies], 'ward')
plt.figure(figsize=(30,20))
plt.title("Dendograma jerarquico para el cluster")
plt.xlabel('ID de los usuarios de Netflix')
plt.ylabel('Distancia')
dendrogram(Z, leaf_rotation=90., leaf_font_size=20.0)
plt.show
```

```python
Z = linkage(data[movies], 'average')
plt.figure(figsize=(30,20))
plt.title("Dendograma jerarquico para el cluster")
plt.xlabel('ID de los usuarios de Netflix')
plt.ylabel('Distancia')
dendrogram(Z, leaf_rotation=90., leaf_font_size=20.0)
plt.show
```

```python
Z = linkage(data[movies], 'complete')
plt.figure(figsize=(30,20))
plt.title("Dendograma jerarquico para el cluster")
plt.xlabel('ID de los usuarios de Netflix')
plt.ylabel('Distancia')
dendrogram(Z, leaf_rotation=90., leaf_font_size=20.0)
plt.show
```

```python
Z = linkage(data[movies], 'single')
plt.figure(figsize=(30,20))
plt.title("Dendograma jerarquico para el cluster")
plt.xlabel('ID de los usuarios de Netflix')
plt.ylabel('Distancia')
dendrogram(Z, leaf_rotation=90., leaf_font_size=20.0)
plt.show
```
