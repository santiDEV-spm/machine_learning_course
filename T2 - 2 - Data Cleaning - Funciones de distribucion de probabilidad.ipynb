# La simulacion de Monte Carlo

* generar dos numeros aleatorios uniformes x e y entre 0 y 1 mil veces
* calcular (x*x) + (y*y)
     * si el valor es inferior a 1 -> estamos dentro del circulo
     * si el valor es superio a 1-> estamos fuera del circulo
 
* calculamos las veces que caemos dentro del circulo y dividimos entre el numero total de intentos para calcular la aproximacion
  de la probabilidad de caer dentro del circulo
 
* Usamos dicha probabilidad para aproximar 

* Repetimos un numero de veces suficiente(por ejemplo 1000), este experimento para obtener multiples aproximaciones de 

* calculamos el promedio del resultado de las repeticiones del experimento para dar la aproximacion final de 

```python
pi_avg = 0
pi_values_list = []
n = 100000
n_experimentos = 300
for i in range(n_experimentos):
    value = 0
    x = np.random.uniform(0,1,n).tolist()
    y = np.random.uniform(0,1,n).tolist()
    
    for j in range(n):
        z = np.sqrt(x[j]*x[j] + y[j] * y[j])
        if(z<=1):
            value += 1
            
    float_value = float(value)
    pi_value = float_value * 4/n
    pi_values_list.append(pi_value)
    pi_avg += pi_value
    
    
pi = pi_avg/n_experimentos

print(pi)

plt.plot(pi_values_list)
        
```
